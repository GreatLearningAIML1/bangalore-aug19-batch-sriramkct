## Question 7

### Define weight and bias
- Initialize weight and bias with tf.zeros
- tf.zeros is an initializer that generates tensors initialized to 0
- Specify the value for shape

```python
W = tf.zeros(shape=(4,1))
b = tf.zeros(shape=(1))
```

## Question 8

### Get prediction
- Define a function to get prediction
- Approach: prediction = (X * W) + b; here is X is features

```python
@tf.function
def predictor(features, W, b):
  y_pred = tf.add(tf.matmul(features, W), b)
  return y_pred
```

### Calculate loss
- Calculate loss using predictions
- Define a function to calculate loss
- We are calculating mean squared error

```python
@tf.function
def loss(y_actual, y_predicted):
  return tf.reduce_mean(tf.square(y_actual - y_predicted))
```

## Question 9

### Define a function to train the model
1.   Record all the mathematical steps to calculate Loss
2.   Calculate Gradients of Loss w.r.t weights and bias
3.   Update Weights and Bias based on gradients and learning rate to minimize loss

```python
def train(X, y_actual, W, b, learning_rate=0.01):
  with tf.GradientTape() as tape:
    tape.watch([W,b])
    curr_prediction = predictor(X, W, b)
    curr_loss = loss(y_actual, curr_prediction)

  #evalute the gradient with the respect to the paramters
  dW, db = tape.gradient(curr_loss, [W, b])

  W = W - learning_rate * dW
  b = b - learning_rate * db

  return W,b
```

## Question 10

### Train the model for 100 epochs 
- Observe the training loss at every iteration

```python
for i in range(100):
  W, b = train(features_l2_norm, labels_l2_norm, W, b)

  if i % 10 == 0:
    print("Step {:03d}: Loss {:.6f}: ".format(i, loss(labels_l2_norm, predictor(features_l2_norm, W, b)).numpy()))

print(f'W : {W.numpy()} , b  = {b.numpy()} ')
```
