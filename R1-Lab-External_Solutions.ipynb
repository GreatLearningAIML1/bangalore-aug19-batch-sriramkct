### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
# IQR calculatio without using Numpy
age_ds = hr_ds['Age']
age_qunatiles = age_ds.quantile([0.25, 0.5, 0.75])
percentile_25 = age_qunatiles[0.25]
percentile_75 = age_qunatiles[0.75]
iqr = percentile_75 - percentile_25
print('IQR for Age : ', iqr)
```

```python
#IQR Calculation using Numpy
perc_25 = np.percentile(age_ds,25)
perc_75 = np.percentile(age_ds,75)
print('IQR for Age(using Numpy) : ',perc_75 - perc_25)
```
