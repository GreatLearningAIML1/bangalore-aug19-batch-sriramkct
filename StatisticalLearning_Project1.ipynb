```python
#1 - Replacing all - with 0
df_sports.replace(to_replace='-',value=0,inplace=True)
df_sports.head(10)
```

```python
df_sports['Points'] = pd.to_numeric(df_sports['Points'])
df_sports['Debut'] = df_sports['Debut'].map(lambda x: re.split('-',x)[0])
df_sports['Debut'] = pd.to_numeric(df_sports['Debut'])
df_sports['GamesWon'] = pd.to_numeric(df_sports['GamesWon'])
df_sports['GamesPlayed'] = pd.to_numeric(df_sports['GamesPlayed'])
df_sports['GoalsFor'] = pd.to_numeric(df_sports['GoalsFor'])
df_sports['GoalsAgainst'] = pd.to_numeric(df_sports['GoalsAgainst'])
```

```python
#2 Teams that had the DEBUT between 1930 ansd 1980
debut_series = df_sports[['Team','Debut']][df_sports.Debut.between(1930,1980, inclusive=True)]
print('Number of teams that had the debut between 1930 and 1980 : ',debut_series['Team'].count())
print("Team names and Debut years are:")
debut_series
```

```python
#3 Teams which came Top 5 in terms of points
print("Top 5 teams by Points : ")
df_sports[['Team','Points']].sort_values(by='Points', ascending=False).head(5)
```

```python
def goal_diff_count(df):
    goal_diff = df.GoalsFor - df.GoalsAgainst
    df['Goal_diff_count'] = goal_diff
```

```python
#4 Finding the goal difference and the team with MAX and MIN goal difference
goal_diff_count(df_sports)
print('Team with MAX goal difference : ', df_sports[['Team']][df_sports.Goal_diff_count == df_sports.Goal_diff_count.max()].iloc[0,0])
print('Team with MIN goal difference : ', df_sports[['Team']][df_sports.Goal_diff_count == df_sports.Goal_diff_count.min()].iloc[0,0])
```

```python
#5 Adding a new column to calculate the WINNING PERCENTAGE for each of the teams
df_sports['Winning Percent'] = (df_sports.GamesWon/df_sports.GamesPlayed) * 100
df_sports.isna().any()
```

```python
df_sports['Winning Percent'].fillna(0,inplace=True)
df_sports.isna().any()
```

```python
print("Top 5 teams by Winning Percentage : ")
df_sports[['Team','Winning Percent']].sort_values(by='Winning Percent', ascending=False).head(5)
```
